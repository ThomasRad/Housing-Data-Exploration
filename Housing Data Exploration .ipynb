```python
#Visually see the distribution of MEDV, the target variable for this dataset. 

sns.histplot(housing_df,x = 'MEDV')
```

```python
# Here we just want to see which variables are pair-wise correlated with the target variable 

corr = housing_df.corr()

#Lets look at the abs corr. 

abs_cor = corr.abs()
abs_cor

#And Now we just want to look at which ones have the highest and filter for those. 

sta = abs_cor.unstack()
so = sta.sort_values(kind="quicksort")
so.MEDV
```

```python
#Lets look at the highest corr variables. We see that there is a negative trend. 

sns.scatterplot(data = housing_df,x='LSTAT', y = 'MEDV')
```

```python
#Now checking for another variable with the target we see that there is a positive trend. 

sns.scatterplot(data = housing_df,x='RM', y = 'MEDV')
```
